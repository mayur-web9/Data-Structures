**Data Types and Structures Questions**


1. What are data structures, and why are they important ?

   => Data structures are ways of organizing data so we can use it easily and efficiently. They matter because they help make programs faster and easier to manage


2. Explain the difference between mutable and immutable data types with examples ?

   => Mutable types can be changed, like a list:
   my_list = [1, 2, 3] (you can update the values).
   Immutable types can't be changed, like a tuple:
   my_tuple = (1, 2, 3) (you can't modify items).


3. What are the main differences between lists and tuples in Python ?

   => Lists can be changed, while tuples can't.
   Lists use square brackets, and tuples use parentheses.
   Lists are used when data needs to be updated whereas tuples are for fixed data.


4. Describe how dictionaries store data.

   => Dictionaries store information as key–value pairs, like labels pointing to specific data. Example: {"name": "Mayur"}.
   Each key is unique, so you can't have two identical keys.


5. Why might you use a set instead of a list in Python ?

   => Sets automatically remove duplicates and allow faster searching, so they're useful when you only need unique items. and They support set operations like union, intersection, and difference, which lists don't handle easily.


6. What is a string in Python, and how is it different from a list

   => A string is simply a sequence of characters, like a word or a sentence. It's written inside quotes.
   The main difference is that a string can't be changed once it's created, but a list can. Also, lists can hold many types of data - numbers, words, even other lists — while a string only holds characters.


7. How do tuples ensure data integrity in Python

   => Tuples are fixed, meaning once you create one, its values cant be changed. This makes them reliable when you want your data to stay the same throughout the program.
   Because nothing can be added, removed, or edited, tuples help prevent accidental changes.


8. What is a hash table, and how does it relate to dictionaries in Python

   => A hash table is a structure that stores data in a way that allows very fast lookups. It uses a unique key to find the value almost instantly. Python dictionaries are built using hash tables behind the scenes, which is why accessing a value by its key is so quick.


9. Can lists contain different data types in Python ?

   => Yes, Python is very flexible with lists.
   You can put numbers, strings, booleans, or even other lists together in one list.
   For example: [10, "hello", 2.5, True] is completely valid.


10. Explain why strings are immutable in Python.

    => Strings are made immutable so Python can handle them more efficiently and safely. Since they don't change, Python can reuse the same string in different places without worrying about it being modified. This also helps avoid unexpected bugs caused by accidental changes.


11. What advantages do dictionaries offer over lists for certain tasks ?

    => Dictionaries let you access data using a key instead of a position, which makes lookups much faster and easier.
    They're great when you need to store related information, like a name and age, because the key clearly describes the value. This makes your code more readable and efficient compared to searching through a list.


12. Describe a scenario where using a tuple would be preferable over a list.

    => A tuple is better when you have data that shouldn’t change, like coordinates (x, y), dates, or configuration settings. Since tuples are fixed, you don’t have to worry about the values being accidentally modified in your program.


13. How do sets handle duplicate values in Python ?

    => Sets simply don't allow duplicates. If you add the same value more than once, the set keeps only one copy.
    This makes sets useful when you want a collection of unique items without manually checking for repeats.
    

14. How does the "in" keyword work differently for lists and dictionaries

    => For a list, "in" checks whether a particular value exists in the list. For a dictionary, "in" checks for keys, not values—so "name" in my_dict looks for a key called "name".


15. Can you modify the elements of a tuple? Explain why or why not

    => No, you can’t change the elements of a tuple.
    Tuples are designed to be immutable, meaning once they're created, their contents are locked. This helps keep certain data safe and prevents accidental changes.


16. What is a nested dictionary, and give an example of its use case

    => A nested dictionary is a dictionary that contains another dictionary inside it.
    It's useful when you want to store structured data, like details about multiple students.
    Example:
    students = {
        "101": {"name": "Mayur", "age": 20},
        "102": {"name": "Aarav", "age": 19}
    }


17. Describe the time complexity of accessing elements in a dictionary

    => Accessing an element in a dictionary is usually O(1), meaning it takes constant time. This is because dictionaries use hashing, which makes lookups very fast.


18. In what situations are lists preferred over dictionaries

    => Lists are preferred when the order of items matters or when you want to store simple collections of data.
    They're great for tasks like looping through items, keeping a sequence, or storing values without needing keys.

19. Why are dictionaries considered unordered, and how does that affect data retrieval

    => Dictionaries don't store items in a fixed position like lists do. Instead, they place items based on their hash values, not index positions. This means you can't rely on the order of elements, but you can always retrieve values quickly using keys.

20. Explain the difference between a list and a dictionary in terms of data retrieval.

    => With a list, you get data by its position (index).
    With a dictionary, you get data by its key, which is faster and more descriptive. Lists work like numbered shelves, while dictionaries work like labeled drawers.
    A list is like finding a book by its shelf number.
    A dictionary is like finding a contact by searching their name on your phone—it goes straight to the person.
